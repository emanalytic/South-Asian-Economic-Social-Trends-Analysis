```python
df.loc[(df['country'] == 'Afghanistan') & (df['year'] == 2023), 'gdp_per_capita_usd'] = 410.93
df.loc[(df['country'] == 'Bhutan') & (df['year'] == 2023), 'gdp_per_capita_usd'] = 3579
```

```python
## in order to interpret easily

df['gdp_current_usd'] = (df['gdp_current_usd'] / 1e9).round(2)  ## GDP in billions
df['population_total'] = (df['population_total'] / 1e6).round(2)  ## Population in millions
df['gdp_per_capita_usd'] = df['gdp_per_capita_usd'].round(0)  ## GDP per capita rounded to nearest whole

df[['gdp_growth_pct', 'unemployment_pct']] = df[['gdp_growth_pct', 'unemployment_pct']].round(2)
```

## Economic Growth




## Average GDP Growth

```python
avg_growth = df.groupby('country')['gdp_growth_pct'].mean().reset_index().sort_values(by='gdp_growth_pct', ascending=False)


def categorize_growth(row):
    if row['gdp_growth_pct'] > 5:
        return 'Stable '
    elif row['gdp_growth_pct'] > 0 and row['gdp_growth_pct'] <= 5:
        return 'Stagnant'
    elif row['gdp_growth_pct'] < 0:
        return 'Negative'
    else:
        return 'Volatile'

avg_growth['Growth Category'] = avg_growth.apply(categorize_growth, axis=1)
avg_growth = avg_growth.rename(columns={
    'country': 'Country',
    'gdp_growth_pct': 'Average GDP Growth (%)',
})
avg_growth
```


## FDI Inflows and Impact on GDP per capita

```python
df_pivot = df.pivot(index='year', columns='country', values='fdi_inflows_pct_gdp')
df_pivot
```

```python
# average fdi % of GDP by country
average_fdi_by_country = df.groupby('country')['fdi_inflows_pct_gdp'].mean().sort_values(ascending=False)
print("Average FDI % of GDP by Country:")
average_fdi_by_country
```

```python
fdi_volatility = df.groupby('country')['fdi_inflows_pct_gdp'].std().sort_values(ascending=False)
print("FDI Volatility (Standard Deviation) by Country:")
print(fdi_volatility)
```

```python
yearly_total_fdi = df.groupby('year')['fdi_inflows_pct_gdp'].sum()

df['Contribution %'] = df.apply(lambda x: (x['fdi_inflows_pct_gdp'] / yearly_total_fdi.loc[x['year']]) * 100, axis=1)
yearly_country_contributions = df.pivot_table(values='Contribution %', index='year', columns='country')
print("Yearly Contribution % of Each Country to Total FDI:")
yearly_country_contributions
```
